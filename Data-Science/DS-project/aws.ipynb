```python
## Remove INR mark in price columns
def removeINR(num):
  val = num[1:]

  return val

## Discounted Price
dp = data.discounted_price.to_list()

dp_new=[]
for i in dp:
  dp_new.append(removeINR(i))

print(dp_new)

data['discounted_price'] = pd.DataFrame(dp_new)

## Actual Price
ar = data.actual_price.to_list()

ar_new =[]

for i in ar:
  ar_new.append(removeINR(i))

print(ar_new)

data['actual_price'] = pd.DataFrame(ar_new)
```

```python
## Remove % mark
## Drop percentage mark
data['discount_percentage'] = data['discount_percentage'].str.replace('\W', '', regex=True)
```

```python
# drop commas in prices
data['discounted_price'] = data['discounted_price'].str.replace('\W', '', regex=True)
```

```python
data['actual_price'] = data['actual_price'].str.replace('\W', '', regex=True)
```

```python
data['rating_count'] = data['rating_count'].str.replace('\W', '', regex=True)
```

```python
## drop NaN rating count rows
Nan_index  = data.index[data['rating_count'].isna()].to_list()
print(Nan_index)
data.drop(Nan_index,inplace=True)
```

```python
import nltk
nltk.download('punkt')
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()

review_val = []
for i in reviews_title:
  review_val.append(sid.polarity_scores(i)['compound'])
```
